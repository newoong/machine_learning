```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Exponential distribution(지수 분포)
#X~Exp(λ) : λ*e^(-λx) (x>=0) / F(x)(누적분포) : 1-e^(-λx)  / E(X)=1/λ / Var(X)=1/λ^2
ld1,ld2,ld3=0.5,1.0,1.5
x=np.linspace(0,3,100)
y1=ld1*np.exp(-ld1*x)
y2=ld2*np.exp(-ld2*x)
y3=ld3*np.exp(-ld3*x)
plt.figure(figsize=(15,8))
plt.plot(x,y1,label='0.5')
plt.plot(x,y2,label='1.0')
plt.plot(x,y3,label='1.5')
plt.legend(loc='best')
```

```python
#누적분포함수
y1=1-np.exp(-ld1*x)
y2=1-np.exp(-ld2*x)
y3=1-np.exp(-ld3*x)
plt.figure(figsize=(15,8))
plt.plot(y1,label='0.5')
plt.plot(y2,label='1.0')
plt.plot(y3,label='1.5')
```

```python
y1=ld1*np.exp(-ld1*x)
y2=ld2*np.exp(-ld2*x)
y3=ld3*np.exp(-ld3*x)

y4=np.random.exponential(1/ld1,1000) #(평균,데이터개수)
y5=np.random.exponential(1/ld2,1000)
y6=np.random.exponential(1/ld3,1000)

plt.figure(figsize=(15,20))
plt.subplot(3,1,1)
plt.hist(y4,bins=300,density=True,alpha=.5)
plt.plot(x,y1)
plt.title('ld=0.5')
plt.xlim(-0.25,3.25)
plt.subplot(3,1,2)
plt.hist(y5,density=True,bins=300,alpha=.5)
plt.plot(x,y2)
plt.title('ld=1.0')
plt.xlim(-0.25,3.25)
plt.subplot(3,1,3)
plt.hist(y6,density=True,bins=300,alpha=.5)
plt.plot(x,y3)
plt.title('ld=1.5')
plt.xlim(-0.25,3.25)
```
